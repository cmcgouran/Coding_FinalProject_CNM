```python
#Import Biopython using the command 'import' in order to access Biopython modules to run a BLAST search. 
import Bio
```

```python
#Import the module NCBIWWW to access the online version of BLAST. 
from Bio.Blast import NCBIWWW
```

```python
#Open the query file (SARS_Cov_2_nt.fasta) by creating a path and printing the contents of the file to verify it is in the correct format (i.e .fasta). 
sequence_data = open('/Users/hvitrbrisingr/Desktop/coding/final_project/NCBI_Blast/bin/SARS_Cov_2_nt.fasta').read() 
print(sequence_data) 
```

```python
#Set the variable 'result_handle' equal to calling the functions of qBLAST and running a nucleotide search over the previously printed sequence data. 
result_handle = NCBIWWW.qblast("blastn", "nt", sequence_data) 
result_handle #should output a io.string# which basically holds the results and comes into play later when parsing through the results. 
```

```python
#Use the 'with open()'' command w/ 'w' mode to write into a new file and write in the query BLAST search findings.
with open('results.xml', 'w') as save_file: 
    blast_results = result_handle.read() 
    save_file.write(blast_results)
```

```python
#Import the NCBIXML module from Biopython Bio.Blast in order to use parsing functions on the output. 
from Bio.Blast import NCBIXML
```

```python
#Parse through the file to only return results that match the secified parameter (i.e. the set E-value). 
E_VALUE_THRESH = 1e-60 #Set E-value equal to a threshold of 1e-60.
for record in NCBIXML.parse(open("results.xml")):            #Use NCBIXML module and apply parse seting the out file as the object.
     if record.alignments:                                   #Use if statement to print the alignment with the following conditions...
        print("\n")                                          #Use print("\n") to print out on seperate lines.
        print("QUERY ACCESSION NUMBER: %s" % record.query[:200]) #Print out showing query(the strings (%S), i.e. the sequences) for each sequence blasted and its top 200 lines of hits. 
        for align in record.alignments: 
           for hsp in align.hsps:                            #Use another for loop to only return the high-scoring alignment regions (HSP) that are less than the E-value specificed. 
            if hsp.expect < E_VALUE_THRESH:                  #Use if statement to draw comparision.     
                print("This is a match to this sequence! Booyah!: %s " % align.title[:200]) #Print the phrase in red and the titles of the top 200 lines of hits.  
             
```
